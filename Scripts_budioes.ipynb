```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
td = pd.read_excel("df2.xlsx", sheet_name="Planilha2")
s = sns.barplot(x="%", y="Genus", data=td, hue="Species")
plt.legend(loc='upper right', fontsize = 8)
legend = plt.legend()
for text in legend.get_texts():
    text.set_fontstyle('italic')
plt.ylabel('')
labels = s.get_yticklabels()
for lbl in labels:
    lbl.set_style('italic')
plt.savefig("Figure3A.png", dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from scipy import stats
%matplotlib inline
graf = pd.read_excel('Duvidas ft FT.xlsx', sheet_name = 'analises')
df_simb = pd.melt(frame=graf,
        id_vars=["Species","Gutpoint","Reads_total_peixe", "Ano","Rtulos de Linha","soma_patogeno"],
        var_name="Genero", 
        value_name="Reads")
df_simb['%'] = df_simb['Reads']/df_simb['Reads_total_peixe']
#sb.barplot(x="%reads", y="Genero", hue="Especie", data=df0, orient="h",errwidth = 0.5)
sb.catplot(x="Gutpoint",y="%", hue="Species",order=['F','MF','M','EM','H'], kind="point", 
            capsize=.15, data=df_simb, aspect=1.5, errwidth=0.5,errorbar = ('sd',0.1),legend=False ).set(title='Symbionts')
data = {'X': ['F','MF','M','EM','H'], 'Y':[0.00068,0.00088,0.0011,0.0011,0.0011]}
data_1 = {'X1': ['F','MF','M','EM','H'], 'Y1':[0.00045,0.0008,0.0012,0.0012,0.0012]}
cts,cts1 = pd.DataFrame(data=data),pd.DataFrame(data=data_1)
sb.lineplot(data=cts, x='X', y='Y',estimator='max', color='red', legend =False)
sb.lineplot(data=cts1, x='X1', y='Y1',estimator='max', color='green', legend =False)
legend = plt.legend()
for text in legend.get_texts():
    text.set_fontstyle('italic')
plt.ylabel('')
#labels = s.get_yticklabels()
#for lbl in labels:
#    lbl.set_style('italic')
#plt.savefig("Figure7A.png", format="svg")
plt.savefig("Figure7A.png", dpi=300, bbox_inches='tight')

plt.show()
```
